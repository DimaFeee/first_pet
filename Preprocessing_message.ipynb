```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import spacy
```

```python
class MessageProcessor:
    def __init__(self):
        self.nlp = spacy.load("en_core_web_sm")
        self.stop_words = set(stopwords.words('english'))

    def preprocess_text(self, text):
        """
        Функция для предобработки текста: удаление специальных символов и приведение к нижнему регистру.
        """
        text = re.sub(r'[^\w\s]', '', text.lower())
        return text

    def remove_stopwords_and_special_chars(self, text):
        """
        Удаление стоп-слов и специальных символов из текста.
        """
        words = word_tokenize(text)
        filtered_words = [word for word in words if word.isalpha()]
        return " ".join(filtered_words)

    def tokenize(self, text):
        """
        Разбиение текста на токены.
        """
        return word_tokenize(text)

    def named_entity_recognition(self, text):
        """
        Определение именованных сущностей в тексте.
        """
        doc = self.nlp(text)
        entities = [(ent.text, ent.label_) for ent in doc.ents]
        return entities
```

```python
# Пример использования
processor = MessageProcessor()
text = "Привет, мир Это пример текста для обработки."
preprocessed_text = processor.preprocess_text(text)
filtered_text = processor.remove_stopwords_and_special_chars(preprocessed_text)
tokens = processor.tokenize(filtered_text)
ner_results = processor.named_entity_recognition(filtered_text)

print(f"Предобработанный текст: {preprocessed_text}")
print(f"Текст после удаления стоп-слов и специальных символов: {filtered_text}")
print(f"Токены: {tokens}")
print(f"Результаты определения именованных сущностей: {ner_results}")
```
